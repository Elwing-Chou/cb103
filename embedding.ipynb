```python
import glob
import pandas as pd
train_df = pd.DataFrame(columns=["content", "sentiment"])
for fname in glob.glob("./aclImdb/train/pos/*.txt"):
    f = open(fname, encoding="utf-8")
    content = f.read()
    f.close()
    s = pd.Series([content, 1], index=["content", "sentiment"])
    train_df = train_df.append(s, ignore_index=True)
for fname in glob.glob("./aclImdb/train/neg/*.txt"):
    f = open(fname, encoding="utf-8")
    content = f.read()
    f.close()
    s = pd.Series([content, 0], index=["content", "sentiment"])
    train_df = train_df.append(s, ignore_index=True)
train_df
```

```python
test_df = pd.DataFrame(columns=["content", "sentiment"])
for fname in glob.glob("./aclImdb/test/pos/*.txt"):
    f = open(fname, encoding="utf-8")
    content = f.read()
    f.close()
    s = pd.Series([content, 1], index=["content", "sentiment"])
    test_df = test_df.append(s, ignore_index=True)
for fname in glob.glob("./aclImdb/test/neg/*.txt"):
    f = open(fname, encoding="utf-8")
    content = f.read()
    f.close()
    s = pd.Series([content, 0], index=["content", "sentiment"])
    test_df = test_df.append(s, ignore_index=True)
test_df
```
