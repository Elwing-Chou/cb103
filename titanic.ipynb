```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("test.csv", encoding="utf-8")
test_df
```

```python
train_df.isna().sum()
```

```python
med = train_df.median()
med
```

```python
train_df = train_df.fillna(med)
```

```python
train_df.isna().sum()
```

```python
frequent = train_df["Embarked"].value_counts().idxmax()
train_df["Embarked"] = train_df["Embarked"].fillna(frequent)
train_df.isna().sum()
```

```python
pd.get_dummies(train_df["Embarked"])
```

```python
dummy = pd.get_dummies(train_df["Embarked"])
train_df = pd.concat([train_df, dummy], axis=1)
```

```python
dummy = pd.get_dummies(train_df["Sex"])
train_df = pd.concat([train_df, dummy], axis=1)
```

```python
train_df
```

```python
def nameflow(n):
    return n.split(",")[1].split(".")[0].replace(" ", "")
ns = train_df["Name"].apply(nameflow)
pd.crosstab(ns, train_df["Survived"]).T
```

```python
def totalnameflow(n):
    title = n.split(",")[1].split(".")[0].replace(" ", "")
    if title == "Mr":
        return "Mr"
    elif title == "Mrs":
        return "Mrs"
    elif title == "Miss":
        return "Miss"
    else:
        return "Drop"
train_df["Name"] = train_df["Name"].apply(totalnameflow)
dummy = pd.get_dummies(train_df["Name"])
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```python
train_df = train_df.drop(["PassengerId", "Name", "Sex", "Ticket", "Cabin", "Embarked", "Drop"], axis=1)
train_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(14, 10))
sns.heatmap(train_df.corr(), annot=True, cmap="RdBu")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np
clf = RandomForestClassifier(n_estimators=27, max_depth=7)
x = train_df.drop(["Survived"], axis=1)
y = train_df["Survived"]
scores = cross_val_score(clf, x, y, cv=10)
print("äº¤å‰é©—è­‰:", scores)
print("å¹³å‡", np.average(scores))
clf.fit(x, y)
```

```python
test_df = test_df.fillna(med)
test_df["Embarked"] = test_df["Embarked"].fillna(frequent)
dummy = pd.get_dummies(test_df["Embarked"])
test_df = pd.concat([test_df, dummy], axis=1)
dummy = pd.get_dummies(test_df["Sex"])
test_df = pd.concat([test_df, dummy], axis=1)
test_df["Name"] = test_df["Name"].apply(totalnameflow)
dummy = pd.get_dummies(test_df["Name"])
test_df = pd.concat([test_df, dummy], axis=1)
test_df = test_df.drop(["Name", "Sex", "Ticket", "Cabin", "Embarked", "Drop"], axis=1)
test_df
```

```python

x = test_df.drop(["PassengerId"], axis=1)
pre = clf.predict(x)
resultdf = pd.DataFrame()
resultdf["PassengerId"] = test_df["PassengerId"]
resultdf["Survived"] = pre
resultdf.to_csv("predict.csv", encoding="utf-8", index=False)
resultdf
```
