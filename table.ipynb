apply -> 把你原本的東西 -> {流程} -> 新的東西

```python
from datetime import datetime
def convert(t):
    return str(datetime.fromtimestamp(t))
df["film_real"] = df["film_date"].apply(convert)
df
```

```python
df.head(3)[ [True, True, False] ]
```

```python
def cfilter(data):
    if data >= 2000 and data < 5000:
        return True
    else:
        return False
s = df["comments"].apply(cfilter)
df[s]
```

```python
def tagflow(data):
    taglist = eval(data)
    if "children" in taglist:
        return True
    else:
        return False
df[ df["tags"].apply(tagflow) ]
```

```python
def test(**kwargs):
    print(kwargs)
test(a = 3, b = 4, c =5)
```

```python
def tagflow(data, **kwargs):
    taglist = eval(data)
    t = kwargs["target"]
    if t in taglist:
        return True
    else:
        return False
df[ df["tags"].apply(tagflow, target="biology") ]
```
